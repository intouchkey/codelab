# SOLUTION

The solution can be easily solved if we use a *priority queue* (min heap) and *greedy method* to solve this problem. If we keep selecting the two smallest numbers, then, we calculate the sum and insert the cost (sum of two min numbers) back into the priority queue and repeat the step until 1 number is left in the heap, by this way, you will always be able to minimize the cost of each summation because the selected numbers to be summed are always the minimum. Therefore, the total cost will be minimal. The reason why a min heap should be used is because when a priority queue is stored in a min heap structure, its structure always maintain min element at the root, so we can easily extract minimum number.

# CODE

````
import heapq
import copy

def solve(input):
    array = input.copy()
    heapq.heapify(array)
    result = 0
    
    while(len(array) > 1):
        cost =  heapq.heappop(array) + heapq.heappop(array)
        result += cost
        heapq.heappush(array, cost)

    return result
````

````
# Test script

for i in range(len(testCases)):
  print("*** Testing program with test case no.", i)

  %timeit solve(testCases[i].input)
  output = solve(testCases[i].input)

  if output==testCases[i].output:
    print("Okay")
  else:
    print("Incorrect!\nExpected output:", testCases[i].output)
    print("Actual output:", output)
````
